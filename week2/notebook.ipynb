## Ejercicio práctico: Entiende una neurona en 15 minutos (Red 'vainilla')

Objetivo: Observar como cambia la salida de una neurona cuando modificas entradas, pesos (w1, w2) y el sesgo (bias).


```python
import numpy as np

def step(z):
    return 1 if z >= 0 else 0

def neurona(x1, x2, w1, w2, b):
    z = x1*w1 + x2*w2 + b
    y = step(z)
    return z, y

# Valores base
w1, w2 = 1, 1
casos = [(0,0),(0,1),(1,0),(1,1)]
```

```python
for b in [-0.5, -1.0, -2.0]:
    print(f"\n=== Probando b = {b} ===")
    for x1, x2 in casos:
        z, y = neurona(x1, x2, w1, w2, b)
        print(f"Entrada=({x1},{x2}) z={z:.1f} salida={y}")
```
