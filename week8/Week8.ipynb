```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import numpy as np

# 1. Carga de imagenes
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Entrenamos 5,000 imágenes 
train_images, train_labels = train_images[:5000], train_labels[:5000]
test_images, test_labels = test_images[:1000], test_labels[:1000]

# Normalización 
train_images, test_images = train_images / 255.0, test_images / 255.0

# 2. CNN BASE SIMPLIFICADA (Menos capas = más rápido)
model_base = models.Sequential([
    layers.Conv2D(16, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(10, activation='softmax')
])

model_base.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

print("Entrenando CNN Base")
model_base.fit(train_images, train_labels, epochs=3, verbose=1) # Solo 3 épocas

# 3. TRANSFER LEARNING 
# Usamos MobileNetV2 pero con 2 épocas
base_model = tf.keras.applications.MobileNetV2(input_shape=(32, 32, 3), include_top=False, weights='imagenet')
base_model.trainable = False

model_tl = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(10, activation='softmax')
])

model_tl.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

print("\nEntrenando Transfer Learning")
model_tl.fit(train_images, train_labels, epochs=2, verbose=1)

# 4. EVALUACIÓN FINAL
loss, acc = model_base.evaluate(test_images, test_labels, verbose=0)
print(f"\nResultado final - Precisión: {acc*100:.2f}%")
```
